## Plotting the model fitted to the TD and BU models

## Testing Varying emission factors
This function takes the species to test as an input. It also takes an estimation of the expected initial consumption and expected growth in consumption anually. It tests a range of emission factors and tests how it will affect the initial consumption required for the model to fit the TD and BU emissions. 

```python

plot = base2.Test_Ef('HFC-134a', 100, 0.05)
plot = base2.Test_Ef('HFC-143a', 10, 0.05)
plot = base2.Test_Ef('HFC-227ea', 10, 0.05)
plot = base2.Test_Ef('HFC-152a', 10, 0.05)
```
